```python
import seaborn as sns
import matplotlib.pyplot as plt

import pandas as pd, numpy as np
```

```python
countries =pd.read_csv('countries-of-the-world.csv')
countries.head()
```

```python
countries.Birthrate = countries.Birthrate.str.replace(',','')
countries.Deathrate = countries.Deathrate.str.replace(',','')
# countries['Birthrate'] = countries['Birthrate'].str.replace(',', '')
countries.head()
```

```python
countries.Birthrate
```

```python
countries[['Deathrate']] = countries[['Deathrate']].fillna(countries[['Deathrate']].mean())
countries[['Birthrate']] = countries[['Birthrate']].fillna(countries[['Birthrate']].mean())

countries.Deathrate = pd.to_numeric(countries.Deathrate, errors='coerce')
countries.Birthrate = pd.to_numeric(countries.Birthrate, errors='coerce')
```

```python
countries.info()
```

### changing wisker
- by default, the wiksers ected to 1.5* the interquertile range
- you can make them to 2.0 * IQR: whis = 2.0
- Show the 5th and 95th percentiles: whis = [5,95]
- show min and max values: whis= [0,100]

```python
g = sns.catplot(x="time", y = 'total_bill',
               data = tips,
               kind = 'box',
               order = ['Dinner',
                       'Lunch'],
               whis = [0,100])
plt.show()
```

```python
g = sns.catplot(x="time", y = 'total_bill',
               data = tips,
               kind = 'box',
               order = ['Dinner',
                       'Lunch'],
               whis = 2.0)
plt.show()
```

# Point plots


to create a point plot, use the __catplot()__ function and specify the name of the 
- __categorical variable__ to put on the __x-axis (x=____),__ 
- the name of the __quantitative variable__ to summarize on the __y-axis (y=____),__ 
- the Pandas DataFrame to use (data=____), and the type of categorical plot (kind="point").

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point")
        
# Show plot
plt.show()
```

### add capzie

```python
# Add caps to the confidence interval
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
           capsize = 0.2 )
        
# Show plot
plt.show()
```

### Remove joins

```python
# Remove the lines joining the points
sns.catplot(x="famrel", y="absences",
			data=student_data,
            kind="point",
            capsize=0.2,
            join = False)
            
# Show plot
plt.show()
```

```python
# Create a point plot with subgroups
sns.catplot(data = student_data, 
             x= 'romantic',
             y = 'absences',
             hue = 'school',
             kind = 'point')

# Show plot
plt.show()
```

### Remove the confidnce interval

```python
# Turn off the confidence intervals for this plot
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci = None)

# Show plot
plt.show()
```

```python
# Import median function from numpy
from numpy import median

# Plot the median number of absences instead of the mean
sns.catplot(x="romantic", y="absences",
			data=student_data,
            kind="point",
            hue="school",
            ci=None,
            estimator = median
            )

# Show plot
plt.show()
```

### rel and cat are faced grid type obects they can have subplot
- g.fig.suptitle("", y = )
- g.set_titles('This is {col_name}') -> subplots titles
- plt.xticks(rotation = 90)
- g.set(xlabel = "New X label",
     ylabel = "New y label")

```python
# Create scatter plot
g = sns.relplot(x="absences", y="G3", 
            data=student_data,
            kind="scatter", 
            )

# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("g3 absence", y = 1.02)

# Show plot
plt.show()
```

```python
g = sns.catplot(x="time", y = 'total_bill',
               data = tips,
               kind = 'box',
               order = ['Dinner',
                       'Lunch'],
               sym = '' ,
               col = 'day')
# Add a title "Car Weight vs. Horsepower"
g.fig.suptitle("Main TITLE", y = 1.02)

g.set_titles('This is {col_name}')
plt.show()
```

### adding axis label

```python
g = sns.catplot(x="time", y = 'total_bill',
               data = tips,
               kind = 'box',
               order = ['Dinner',
                       'Lunch'],
               sym = '' )

g.set(xlabel = "New X label",
     ylabel = "New y label")

plt.show()
```

### rotating the x axis ticks

- plt.xticks(rotation = 90) works with both faced grid and axis subplot object

```python
g = sns.catplot(x="time", y = 'total_bill',
               data = tips,
               kind = 'box',
               order = ['Dinner',
                       'Lunch'],
               sym = '' ,
               )
plt.xticks(rotation = 90)
plt.show()
```
